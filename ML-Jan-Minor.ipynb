# Analysis based on Size of Game

```python
df4=pd.DataFrame(df2)
print("The game below 2 stars rating having the maximum size is:")
df4.loc[[df4['Size'].argmax()]]
```

So we can observe that games that are action,simulation & strategy oriented occupy higher space.

# Analysis based on Price Tag

```python
df3=pd.DataFrame(df1)
print("The game above 4 stars having the highest price tag is:")
df3.loc[[df3['Price'].argmax()]]
```

Thus,we can find out that simulation,finance and strategy games with average user ratings above 4 stars have more demand and users are ready to purchase them by investing their money.

```python
print("The game below 2 stars rating with the highest price tag is:")
df4.loc[[df4['Price'].argmax()]]
```

We can see that users are not ready to invest their money in poor rated games. Hence,the game above 4 stars has a price rating of 139.99 and an user rating count of 1099.0 whereas the game below 2 stars has a price of just 9.99 and an user rating count of just 7.0. Thus we can clearly observe the visible difference between them with the help of the data.

# Analysis based on User Rating Count

```python
print("The game above 4 stars rating with the highest user rating count is:")
df3.loc[[df3['User Rating Count'].argmax()]]
```

Here we can observe that user rating count is higher when the game is available for free in the appstore as well has a good average user ratings favourably 4+ stars.

```python
print("The game below 2 stars rating with the highest user rating count is:")
df4.loc[[df4['User Rating Count'].argmax()]]
```

This piece of data clearly shows the importance and dependency of user rating count on average user rating. The game Clash of Clans having an average user rating of 4.5 has an user rating count of 3032734.0 but the game Parking Escape having an average user rating of mere 1.5 has an user rating count of only 1047.0. Therefore,we can see a clear cut difference upon analyzing the data. 

INFERENCE:

1)The main conclusion is that games which are freely available in the app store tend to have an higher user rating count.

2)And games that are actiuon and entertainment oriented with good average user ratings have higher user rating count.

3)And games that have more newer versions and are being updated more frequently tend to have higher user rating count. 

# Analysis based on In-app Purchases: 

When we observe the above collection of data, we can note that the games which have higher average user ratings and frequent updates have witnessed more in-app purchases

# Graphical Analysis:

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
dg=pd.DataFrame(df)
plt.plot(dg['Average User Rating'],dg['Price'])
plt.xlabel('Average User Rating')
plt.ylabel('Price')
plt.title('Dependency of user ratings on price')
plt.show()
```

INFERENCE:

Paid games have better average user ratings than the free ones.

```python
plt.plot(dg['User Rating Count'],dg['Average User Rating'])
plt.xlabel('User Rating Count')
plt.ylabel('Average User Rating')
plt.title('Dependency of Average User Rating on User Rating Count')
plt.show()
```

INFERENCE:

Higher rated apps have higher user rating count.
